## B) Data for Classification

For binary classification ie class 0 or 1

```python
u1 = np.array([2,4])
cov_1 = np.array([[1,0],
                 [0,1]])

u2 = np.array([-1,-3])
cov_2 = np.array([[1,0],
                 [0,2]])
```

```python
x1 = np.random.multivariate_normal(u1,cov_1,300)
print(x1.shape)

x2 = np.random.multivariate_normal(u2,cov_2,400)
print(x2.shape)
```

```python
plt.scatter(x1[:,0],x1[:,1],label="Apple")
plt.scatter(x2[:,0],x2[:,1],label="Orange")
plt.xlabel("x1-Size")
plt.ylabel("x2-Sweetnness")
plt.legend()
plt.show()
```

```python
y = np.zeros(700)
print(y.shape)
```

```python
y[:300]=1 #first 300 rows 1
print(y)
```

```python
X = np.vstack((x1,x2))
print(X.shape)
```

```python
#plt.scatter(X[:,0],X[:,1]) Both of same color difficult to differentiate

for i in range(700):
    if y[i]==1 :
        plt.scatter(X[i,0],X[i,1],c="green")
    else:
        plt.scatter(X[i,0],X[i,1],c="orange")

plt.show()
```

# Shortcut

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```
