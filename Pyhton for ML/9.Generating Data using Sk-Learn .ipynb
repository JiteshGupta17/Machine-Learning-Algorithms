```python
from sklearn.datasets import make_classification
```

```python
make_classification?
```

```python
Xc,Yc = make_classification(n_samples=400,n_features=2,n_informative=2,n_classes=3,n_clusters_per_class=1,n_redundant=0,random_state=5)
```

```python
print(Xc.shape,Yc.shape)
```

```python
plt.scatter(Xc[:,0],Xc[:,1],c=Yc)
plt.show()
```
