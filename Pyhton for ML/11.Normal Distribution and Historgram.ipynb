## Normal Distribution(1-d)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# For standard normal , which is given by randn
# u = 0
# sigma = 1

u = 5
sigma = 1

vals =u + sigma*np.random.randn(1000)
print(vals.shape)
#print(vals)

plt.hist(vals,50)
plt.show()
```

```python
valss = np.round(vals)
# print(vals)

z = np.unique(valss,return_counts=True)
print(z)
```

```python
x = vals
y = np.zeros(x.shape)


plt.scatter(x,y) 
```

# MultiVariate Normal Distribution

```python
# BIvariates
mean = np.array([0.0,0.0])
cov = np.array([[1,0.8],
               [0.8,1]])



dist = np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
```

```python
plt.scatter(dist[:,0],dist[:,1])
```

```python
mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,0.2],
               [0.2,1.1]])

dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(dist2.shape)

plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
```
